```python
import cv2 as cv
im=cv.imread(r'gal_gaussian.png')

blur= cv.GaussianBlur(im,(5,5),0)

cv.namedWindow('Image',cv.WINDOW_AUTOSIZE)
cv.imshow('Image',im)
cv.waitKey(0)
cv.imshow('Image',blur)
cv.waitKey(0)
cv.destroyAllWindows()
```

```python
import cv2 as cv
img=cv.imread(r'gal_sandp.png')
saltnppr= cv.medianBlur(img,5,3)

cv.namedWindow('Image',cv.WINDOW_AUTOSIZE)
cv.imshow('Image',img)
cv.waitKey(0)
cv.imshow('Image',saltnppr)
cv.waitKey(0)
cv.destroyAllWindows()
```

```python
import numpy as np
import cv2 as cv

im=np.zeros((40,60),dtype=np.uint8)

im[0:21,30:61]=125
cv.imshow('image',cv.WINDOW_AUTOSIZE)
cv.imshow('Image',im)
cv.waitKey(0)
cv.destroyAllWindows()
```

```python
import numpy as np
import cv2 as cv
import matplotlib.pyplot as plt

im=np.zeros((40,60),dtype=np.uint8)

im[0:21,30:61]=125

fig,ax = plt.subplots()
ax.imshow(im,cmap='gray',vmin=0,vmax=255)
plt.show()
```

```python
%matplotlib inline
import numpy as np
import cv2 as cv
import matplotlib.pyplot as plt
img = cv.imread('tom_dark.jpg', cv.IMREAD_COLOR)
img = cv.cvtColor(img, cv.COLOR_BGR2RGB)
fig, ax= plt.subplots()
img2=img+75
ax.imshow(img2)
plt.show()
```

```python
import numpy as np
import cv2 as cv
import matplotlib.pyplot as plt

f=cv.imread(r'gal_gaussian.png',cv.IMREAD_GRAYSCALE)
assert f is not None

t= np.arange(255,-1,-1,dtype=np.uint8)
g=cv.LUT(f,t) # g=t[f]

fig,ax=plt.subplots()
ax.plot(t) 


cv.namedWindow('Image',cv.WINDOW_AUTOSIZE)
cv.imshow('Image',f)
cv.waitKey(0)
cv.imshow('Image',g)
cv.waitKey(0)
cv.destroyAllWindows()
```
